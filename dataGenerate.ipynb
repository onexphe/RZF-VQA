```python
def testByJSON(data):
    with open('./test.json','w',encoding='utf8') as f:
        json.dump(data, f, sort_keys=True, indent=4, separators=(',', ':'), ensure_ascii=False)
```

```python
#数据增广

with open('./data/KG_VQA/fvqa/exp_data/train_data/train3/all_qs_dict_release_train_500.json','r',encoding='utf8') as f:
    train_data = json.load(f)
with open('./data/KG_VQA/fvqa/exp_data/common_data/answer.vocab.fvqa.fact.500.json','r',encoding='utf8') as f:
    fact_dict = json.load(f)


allQ = {
    'N':[],
    'J':[],
    'V':[],
    'NUM':[],
    'IN':[],
    'F':[],
}
allQ_err =[]

allType_entity = []

change = {
    'NN':'N',
    'JJ':'J',
    'VBN':'V',
    'VBG':'V',
    'VB':'V',
    'CD':'NUM',
    'DT':'N',
    'NNS':'N',
    'RB':'F',
    'IN':'IN',
    'VBP':'V',
    'PRP':'N',
    'PRP$':'N',
    'RBR':'J',
    'JJS':'J',
    'WP':'N',
    'WRB':'F',
    'MD':'V',
    'VBD':'V',
    'RBS':'J'
}
for key,value in train_data.items():
    fact = value['fact']
    question = value['question']
    answer = value['answer']
    if fact[0] == answer:
        entity = fact[2]
    if fact[2] == answer:
        entity = fact[0]
    result = pos_tag(word_tokenize(entity))
    if result[0][1] not in allType_entity:
        allType_entity.append(result[0][1])
    if entity in question:
        type_e = change[result[0][1]]
        temp = question.replace(entity,'<%s>'%type_e)
        if temp not in allQ[type_e]:
            allQ[type_e].append(temp)
            #allQ.append([temp,entity])
    else:
        allQ_err.append(question)
for key,value in allQ.items():
    print("%s:%s"%(key,value))
    
testByJSON(allQ)
print(len(train_data))
print(len(allQ))
print(len(allQ_err))

print(allType_entity)

    
```
