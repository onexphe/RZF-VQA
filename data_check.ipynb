```python
word = "He was elected president in 2008."
max_seq_length = 30
# tokenizer = BertTokenizer.from_pretrained()
tokenizer = BertTokenizer('./model/re_agcn/vocab.txt')
tokens = []
token = tokenizer.tokenize("He was elected president in 2008.")
tokens.extend(token)

input_ids = tokenizer.convert_tokens_to_ids(tokens)
input_mask = [1] * len(input_ids)
segment_ids = [0] * len(tokens)

padding = [0] * (max_seq_length - len(input_ids))
input_ids += padding
input_mask += padding
segment_ids += padding

assert len(input_ids) == max_seq_length
assert len(input_mask) == max_seq_length
assert len(segment_ids) == max_seq_length


config = BertConfig.from_json_file(os.path.join("./model/re_agcn/bert_config.json"))
model = BertModel(config)
# model = model.from_pretrained('bert-base-uncased')

input_ids = torch.tensor(input_ids, dtype=torch.long)
input_mask = torch.tensor(input_mask, dtype=torch.long)
# valid_ids = torch.tensor(self.data[index]["valid_ids"], dtype=torch.long)
segment_ids = torch.tensor(segment_ids, dtype=torch.long)


input_ids = input_ids.view(1,-1)
input_mask = input_mask.view(1,-1)
segment_ids = segment_ids.view(1,-1)


sequence_output, pooled_output = model(input_ids, segment_ids, input_mask, output_all_encoded_layers=False)

print(input_ids)
print(token)
print(sequence_output.shape,pooled_output.shape)
```

# 数据分布测试

```python
#测试集fact分布 data5
import os ,json
dic = {}
with open('./data/KG_VQA/fvqa/exp_data/test_data/test5/all_qs_dict_release_test_500.json','r',encoding='utf8') as f:
    data = json.load(f)
    print(len(data))
    for key,value in data.items():
        fact = value['fact']
        answer = value['answer']
        question = value['question']
        if fact[0] == answer:
            fact1 = fact[2]
        if fact[2] == answer:
            fact1 = fact[0]
        dic[str(fact1)] = dic[str(fact1)]+1 if str(fact1) in dic else 1
dic = sorted(dic.items(), key= lambda d:d[1], reverse=False)  
print(dic)
```
